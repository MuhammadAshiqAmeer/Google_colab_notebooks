```python
import io
import pandas as pd
```

```python
df=pd.read_csv(io.BytesIO(uploaded['adult.data.csv']))
```

```python
# How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
rc=df['race']
d=rc.value_counts().tolist()
d
```

```python
# What is the average age of men?
avg=round(df[df['sex'] == 'Male']['age'].mean(),1)


avg

```

```python
# What is the percentage of people who have a Bachelor's degree?
bat=(df['education']=='Bachelors').value_counts()
bat[1]/(bat[0]+bat[1])

```

```python
 # What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
# What percentage of people without advanced education make more than 50K?
# with and without `Bachelors`, `Masters`, or `Doctorate`
# percentage with salary >50K
hed=df[df['education'].isin(['Bachelors','Masters','Doctorate'])]
higher_education_rich=round((len(hed[hed['salary'] == '>50K'])/len(hed)*100),1)
higher_education_rich

led=df[~df['education'].isin(['Bachelors','Masters','Doctorate'])]
lower_education_rich=round((len(led[led['salary'] == '>50K'])/len(led)*100),1)
lower_education_rich
                              
                      
```

```python

# What is the minimum number of hours a person works per week (hours-per-week feature)?
m=df['hours-per-week'].min()

```

```python
#What percentage of the people who work the minimum number of hours per week have a salary of >50K?
r=round((len(df[(df['hours-per-week'] == m) & (df['salary'] == '>50K')])/len(df[df['hours-per-week'] == m]))*100,1)
r
```

```python
# What country has the highest percentage of people that earn >50K?
ric=df[df['salary']=='>50K']['native-country'].value_counts()
c=df['native-country'].value_counts()
richest=(ric/c).max()
rc=(ric/c).sort_values(ascending=False).index[0]
rc
```

```python
# Identify the most popular occupation for those who earn >50K in India.
india=df.loc[(df['salary']=='>50K') & (df['native-country']=='India')]['occupation']
india.mode()[0]

```
