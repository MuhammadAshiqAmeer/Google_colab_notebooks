```python
#1. Read an array of numbers. Read two numbers a1 and a2. Prints the elements in the array
#that are integer multiple of either a1 or a2 but not both.
#Eg:
#Input Array -> [32, 17, 25, 40, 120, 23, 800, 400, 75, 63, 8]
#a1 -> 5
#a2 -> 8
#Answer -> [ 32, 25, 75, 8]

lst = input("Enter elements: ").split()
ls1 = []
res = []
for i in range(len(lst)):
    ls1.append(int(lst[i]))
a1 = int(input("enter  first no : "))
a2 = int(input("enter second no : "))
for i in range(len(ls1)):
    if ((ls1[i] % a1 == 0) & (ls1[i] % a2 != 0)) | ((ls1[i] % a1 != 0) & (ls1[i] % a2 == 0)):
        res.append(ls1[i])
print(res)


```

```python
#2. Read an array of numbers. If the ith element is ood number, repeat it (i-1) times in the array.
#Print the modified array, size and sum of its elements.
#Eg:
#Input array -> [2, 5, 7, 14, 3]
#Modified array -> [ 2, 5, 5, 7, 7, 7, 14, 3, 3, 3, 3, 3]
#Size -> 12
#Sum -> 62

inp = input("Enter elements : ").split()
arr = []
res = []
for i in range(len(inp)):
    arr.append(int(inp[i]))


print(arr)
c = 0
for i in range(len(arr)):
    l = i+1
    if arr[i] % 2 == 0:
        res.insert(c, arr[i])
        c = c+1
    else:
        while l > 0:
            res.insert(c, arr[i])
            c = c + 1
            l = l - 1


print(res)
print(f"size = {len(res)}")
print(f"sum = {sum(res)}")
```

```python
#3. Read a positive interger. Print all perfect suqares and perfect cubes within that number.
#(without using library function calls)
#Eg:
#Input Number :150
#Perfect square and cubes below 150 are
#[1, 4, 8, 9, 16, 25, 27, 36, 49, 64, 81, 100, 121, 125, 144]

n = int(input("Enter number : "))
res = []
i = 1
while i * i < n:
    res.append(i * i)
    if (i * i * i < n) & (i != 1):
        if i * i * i not in res:
            res.append(i * i * i)
    i += 1
res.sort()
print(f"perfect squares and cubes below {n} are : {res}")
```
