```python
class tree:
  def __init__(self,nodes,left=None,right=None):
    self.nodes=nodes
    self.left=left
    self.right=right


#to find spliting point
def findSplitPoint(nodes):
    arr=[]
    if len(nodes)==1:
      return 0
    for i in range(len(nodes)):
      suml=0
      sumr=0
      if i==0:
        left=int(nodes[i][1])
      else:
        for l in range(0,i+1):
          suml+=nodes[l][1]
        left=suml
      for r in range(i+1,len(nodes)):
        sumr+=nodes[r][1]
      right=sumr
      arr.append(abs(left-right))

    return arr.index((min(arr)))+1


# to store codes of each symbols
def storeCodes(codetree, val=''):
  newVal=val
  if(type(codetree)!=list):
      if(codetree.left):
        newVal = val + '0'
        storeCodes(codetree.left, newVal)
      if(codetree.right):
          newVal=val + '1'
          storeCodes(codetree.right, newVal)

  if(type(codetree)==list):
      codes.append((codetree[0][0],newVal))


#generate coding tree
def shannonCoding(nodesLis,flag=None,move=None):
    #create coding tree
    if len(nodesLis)>1:
      split=findSplitPoint(nodesLis)
  
      newNode=tree(nodesLis)

      rightTree=nodesLis[split:]
      leftTree=nodesLis[:split]

      newNode.left=leftTree
      newNode.right=rightTree

      if len(codeTree)==0:
        codeTree.append(newNode)
        move=codeTree[0]
    
      if flag==0:
        move.left=newNode 
        move=newNode

      if flag==1:
        move.right=newNode
        move=newNode
       
      shannonCoding(move.left,0,move)
      shannonCoding(move.right,1,move)


#decode coded string
def decode(codetree,txt):
    orgTxt=''
    path = codetree
    
    for i in txt:
      if i == '0':
        if(type(path)!=list):
          path = path.left
        if(type(path)==list):
          orgTxt = orgTxt + path[0][0]
          path = codetree

      if i == '1':
        if(type(path)!=list):
          path = path.right 
        if(type(path)==list):
          orgTxt = orgTxt + path[0][0]
          path = codetree
      
    return orgTxt  

#print coded string corresponding to input
def printCodedString():
    codedString=''
    for i in textToCode:
      for a in range(len(codes)):
        if i== codes[a][0]:
          codedString=codedString + codes[a][1] 
    return codedString
```
