# Huffmann coding


```python
class node:
  def __init__(self, symbol,freq, left=None, right=None):

    self.symbol=symbol
    self.freq=freq
    self.left=left
    self.right=right
    self.huff=''



def storeCodes(node, val=''):
    newVal = val + str(node.huff)
    if(node.left):
        storeCodes(node.left, newVal)
    if(node.right):
        storeCodes(node.right, newVal)
    if(not node.left and not node.right):
        codes.append((str(node.symbol),newVal))


def printCodedString():
    codedString=''
    for i in textToCode:
      for a in range(len(codes)):
        if i== codes[a][0]:
          codedString=codedString + codes[a][1] 
    return codedString


def decode(treeRoot,txt):
    orgTxt=''
    path = treeRoot
    for i in txt:
      if i == '0':
        path = path.left
      if i == '1':
        path = path.right 
      if (not path.left and not path.right):
        orgTxt = orgTxt + path.symbol
        path = treeRoot
    return orgTxt

    

#------------------------------execution---------------------------------------------
codes=[]
frequency={}
nodes=[]

#------------------------coding-------------------------------------

textToCode = input("Enter Text to code using Huffmann code :")

#insert symbols and frequency into dictionary
for i in textToCode:
  frequency[i]=textToCode.count(i)
  
symbol=list(frequency.keys())
freq=list(frequency.values())

# symbol and frequency as node objects
for x in range(len(symbol)):
  nodes.append(node(symbol[x],freq[x]))

#create huffmann coding tree
while len(nodes)>1:
  #sort nodes in ascending order
  nodes = sorted(nodes, key=lambda x: x.freq)

  left=nodes[0]
  right=nodes[1]

  left.huff=0
  right.huff=1

  newNode=node(left.symbol+right.symbol,left.freq+right.freq,left,right)
  nodes.remove(left)
  nodes.remove(right)
  
  nodes.append(newNode)

#display outputs
storeCodes(nodes[0])
print(f"codes are : {codes}")

print(f"coded string : {printCodedString()}")

#-----------decoding----------------------------------------

print(f"decoded string : {decode(nodes[0],printCodedString())}")
```
